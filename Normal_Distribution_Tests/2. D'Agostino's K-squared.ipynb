<h3><font color='grey'>Checks normality based on skewness & kurtosis</font></h3>

<ul>
    <li><code>Skewness</code> - Check for symmetry.</li>
    <br>
    <li><code>Kurtosis</code> - Check for tails.</li>
</ul>

```python
# libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# style
from ipynb.fs.full.i_Style_Sheet import *
plt.style.use(style_use)
plt.rcParams.update(style_params)
```

```python
# scipy
from scipy.stats import normaltest

# random normal
np.random.seed(1)
mu, sigma, s = 10, 1, 10000
norm = np.random.normal(mu, sigma, s)
stat, p = normaltest(norm)

# logic
if p > 0.05:
    a = f'P-value = {round(p, 4)} >> Probably Gaussian'
else:
    a = f'P-value = {round(p, 4)} >> Probably not Gaussian'

# show
plt.hist(norm, density=True, alpha=0.5)
plt.title(f'{a}')
plt.show()
```

```python
# scipy
from scipy.stats import normaltest

# random poisson
np.random.seed(1)
mu, sigma, s = 10, 1, 10000
pois = np.random.poisson(mu, s)
stat, p = normaltest(pois)

# logic
if p > 0.05:
    a = f'P-value = {round(p, 4)} >> Probably Gaussian'
else:
    a = f'P-value = {round(p, 4)} >> Probably not Gaussian'

# show
plt.hist(pois, density=True, alpha=0.5)
plt.title(f'{a}')
plt.show()
```
