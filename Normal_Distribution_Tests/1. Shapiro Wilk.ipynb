```python
# libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# style
from ipynb.fs.full.i_Style_Sheet import *
plt.style.use(style_use)
plt.rcParams.update(style_params)
```

```python
# scipy
from scipy.stats import shapiro

# random normal
np.random.seed(1)
mu, sigma, s = 10, 1, 4999
nml = np.random.normal(mu, sigma, s)
stat, p = shapiro(nml)

# logic
if p > 0.05:
    a = f'P-value = {round(p, 4)} >> Probably Gaussian'
else:
    a = f'P-value = {round(p, 4)} >> Probably not Gaussian'

# show
sns.histplot(data=nml, kde=True, stat='probability', element='step', bins=10, alpha=0.2)
plt.title(f'{a}')
plt.show()
```

```python
# scipy
from scipy.stats import shapiro

# random poisson
np.random.seed(1)
mu, sigma, s = 10, 1, 4999
psn = np.random.poisson(mu, s)
stat, p = shapiro(psn)

# logic
if p > 0.05:
    a = f'P-value = {round(p, 4)} >> Probably Gaussian'
else:
    a = f'P-value = {round(p, 4)} >> Probably not Gaussian'

# show
sns.histplot(data=psn, kde=True, stat='probability', element='step', bins=10, alpha=0.2)
plt.title(f'{a}')
plt.show()
```
